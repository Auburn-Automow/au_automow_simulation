```python
def extract_hough(msg,timestamp):    scan = Scan.from_LaserScan(msg)    xy = scan.ranges[3:,:]    (minx, miny) = np.min(xy,1)    (maxx, maxy) = np.max(xy,1)        range_x = np.ceil((maxx - minx) * 1./RESOLUTION) + 1    range_y = np.ceil((maxy - miny) * 1./RESOLUTION) + 1    np_src = np.zeros((range_x,range_y), dtype=np.uint8)    for el in xy.T:        el_x = np.ceil((el[0] - minx) * 1./RESOLUTION)        el_y = np.ceil((el[1] - miny) * 1./RESOLUTION)        np_src[el_x,el_y] = 255    src = cv.fromarray(np_src)    color_dst = cv.CreateImage(cv.GetSize(src), 8, 3)    cv.CvtColor(src, color_dst, cv.CV_GRAY2BGR)    storage = cv.CreateMemStorage(0)    lines = 0    lines = cv.HoughLines2(src, storage, cv.CV_HOUGH_PROBABILISTIC, 5 , pi/180, 50, 30, 30)    return np_src, lines    
```
