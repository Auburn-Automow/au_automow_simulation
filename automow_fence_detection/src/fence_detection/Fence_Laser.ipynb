```python
from matplotlib.lines import Line2Ddef laser_errorbar(ax, scan, n_devs, plotrange=None):    if plotrange is None:        plotrange = [0, len(scan)]    for ii in np.arange(plotrange[0],plotrange[1],1):        barlen = scan.ranges[2,ii] * n_devs        newline = Line2D([scan.ranges[0,ii],scan.ranges[0,ii]], [scan.ranges[1,ii] - barlen, scan.ranges[1,ii] + barlen], lw=1)        ax.add_line(newline)
```

```python
def segment_scan(scan, threshold, min_cluster):    clusters = []    clusterBegin = 0        for ii in range(len(scan)-1):        r1 = scan.ranges[1,ii]        r2 = scan.ranges[1,ii+1]        if (abs(r1-r2) >= threshold or ii == len(scan)-2 ):            if (ii - clusterBegin) > min_cluster:                clusters.append(scan.get_partition_by_index(clusterBegin, ii))                clusterBegin = ii + 1            else:                clusterBegin = ii +1    return clusters
```

```python
def perpendicular_distance( point, line ):    # Line is given in [ x1, y1; x2, y2 ]    # Point is given in [ x1, y1 ];    m = (line[1,1] - line[0,1]) / (line[1,0] - line[0,0])    b = - m * line[0,0] + line[0,1]    return abs(m * point[0] - point[1] + b)/np.sqrt(m**2  + 1)
```

```python
def douglas_peucker( scan, epsilon):    dmax = 0    index = 0    for i in range(2,len(scan)-2):        d = perpendicular_distance(scan.ranges[3:,i],                                   np.vstack((scan.ranges[3:,0], scan.ranges[3:,-1])))        if d > dmax:            index = i            dmax = d        results = []    if dmax >= epsilon:        recResults1 = douglas_peucker(scan.get_partition_by_index(0,index), epsilon)        recResults2 = douglas_peucker(scan.get_partition_by_index(index,len(scan)-1), epsilon)        if recResults1:            results.extend(recResults1)        if recResults2:            results.extend(recResults2)    else:        results.append(Line2D([ scan.ranges[3,0], scan.ranges[3, -1] ],                              [ scan.ranges[4,0], scan.ranges[4, -1] ], lw=2, color='red'))    return results
```
